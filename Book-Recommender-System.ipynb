Popularity based recommender system

```python
ratings_per_name = ratings.merge(books,on='ISBN')
```

```python
num_ratings_per_book = ratings_per_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
avg_rating_per_book = ratings_per_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
```

```python
final = num_ratings_per_book.merge(avg_rating_per_book,on='Book-Title').rename(columns={'Book-Rating_x':'num_ratings','Book-Rating_y':'avg_rating'}).sort_values('num_ratings',ascending=False)
```

```python
final.head(50).sort_values('avg_rating',ascending=False)
```

Colaboration filtering based recommender system

```python
ratings_per_name
```

```python
a = ratings_per_name.groupby('User-ID').count()['Book-Title'] > 200
```

```python
True_users = a[a].index
```

```python
filtered_df = ratings_per_name[ratings_per_name['User-ID'].isin(True_users)]
```

```python
b = filtered_df.groupby('Book-Title').count()['Book-Rating']>50
```

```python
famous_book = b[b].index
```

```python
final = filtered_df[filtered_df['Book-Title'].isin(famous_book)]
```

```python
final
```

```python
pt = final.pivot_table(columns='User-ID',index='Book-Title',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(pt)
```

```python
sorted(list(enumerate(similarity[0])),key = lambda x:x[1],reverse=True)[1:6]
```

```python
np.where(pt.index=='1984')[0][0]
```

```python
def recommender(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity[index])),key = lambda x:x[1],reverse=True)[1:6]
    
    for book in similar_books:
        print(pt.index[book[0]])
```

```python
recommender('Year of Wonders')
```
