POPULARITY BASED RECOMMENDER : 
HERE WE'LL EXTRACT THE MOST POPULAR 50 BOOKS ON THE BASICS OF ANALYSIS OF DATA AND WILL RECOMMEND THESE 50 BOOKS TO ALL OUR USERS WHO VISIT OUR WEBSITE

```python
books_with_ratings = books.merge(ratings,on='ISBN')
```

```python
books_with_ratings.head()
```

```python
num_of_ratings_per_book = books_with_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
avg_ratings_per_book = books_with_ratings.groupby('Book-Title').mean()['Book-Rating'].reset_index()
```

```python
top_50 = num_of_ratings_per_book.merge(avg_ratings_per_book,on='Book-Title')
```

```python
top_50.head()
```

```python
top_50 = top_50.rename(columns={'Book-Rating_x':'num_ratings','Book-Rating_y':'avg_rating'})
```

```python
top_50.head()
```

```python
final_50 = top_50.sort_values('num_ratings',ascending=False).head(50)
```

```python
final_50.head()
```

```python
final_50.sort_values('avg_rating',ascending=False)
```

COLABARATIVE FILTERING BASED RECOMMENDER

```python
books_with_ratings.head()
```

```python
a = books_with_ratings.groupby('User-ID').count()['Book-Rating']>200
```

```python
True_users = a[a].index
```

```python
filtered_df = books_with_ratings[books_with_ratings['User-ID'].isin(True_users)]
```

```python
b = filtered_df.groupby('Book-Title').count()['Book-Rating'] > 50
```

```python
Good_books = b[b].index
```

```python
filtered_df = filtered_df[filtered_df['Book-Title'].isin(Good_books)]
```

```python
filtered_df
```

```python
pivot = filtered_df.pivot_table(columns='User-ID',index='Book-Title',values='Book-Rating')
```

```python
pivot.head()
```

```python
pivot.fillna(0,inplace=True)
```

```python
pivot
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(pivot)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),key = lambda x:x[1],reverse=True)[1:6]
```

```python
def recommender(book_name):
    index_in_pivot = np.where(pivot.index == book_name)[0][0]
    similarity_score = sorted(list(enumerate(similarity[index_in_pivot])),key = lambda x:x[1],reverse=True)[1:6]
    
    for book in similarity_score:
        print(pivot.index[book[0]])
```

```python
np.where(pivot.index == '1984')[0][0]
```

```python
recommender('1984')
```

```python
recommender('2nd Chance')
```
